```python
import pandas as pd
```

```python
train_X = pd.read_csv('../data/curated/cleaned_train_X.csv')
val_X = pd.read_csv('../data/curated/cleaned_val_X.csv')
test_X = pd.read_csv('../data/curated/cleaned_test_X.csv')
```

## One-hot Encode Categorical Data

```python
encoded_train_X = pd.get_dummies(train_X, columns=['Sex', 'Embarked', 'ticket_prefix', 'title', 'family_name'])
encoded_val_X = pd.get_dummies(val_X, columns=['Sex', 'Embarked', 'ticket_prefix', 'title', 'family_name'])
encoded_test_X = pd.get_dummies(test_X, columns=['Sex', 'Embarked', 'ticket_prefix', 'title', 'family_name'])
```

### PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
def get_pca(encoded_train_X):
    scaler = StandardScaler()
    scaled_train_X = scaler.fit_transform(encoded_train_X)
    pca = PCA(n_components=16)
    principal_components = pca.fit_transform(scaled_train_X)
    principal_df = pd.DataFrame(data=principal_components)
    return principal_df
```

```python
principal_train_df = get_pca(encoded_train_X)
principal_val_df = get_pca(encoded_val_X)
principal_test_df = get_pca(encoded_test_X)
```

```python
principal_train_df.to_csv('../data/curated/final_train_X.csv',index=False)
principal_val_df.to_csv('../data/curated/final_val_X.csv', index=False)
principal_test_df.to_csv('../data/curated/final_test_X.csv', index=False)
```
