```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
train_X = pd.read_csv('../data/curated/cleaned_train_X.csv')
val_X = pd.read_csv('../data/curated/cleaned_val_X.csv')
test_X = pd.read_csv('../data/curated/cleaned_test_X.csv')
```

```python
dropped_train_X = train_X.drop(columns=['family_name'])
dropped_val_X = val_X.drop(columns=['family_name'])
dropped_test_X = test_X.drop(columns=['family_name'])
```

## One-hot Encode Categorical Data

```python
all_X = pd.concat([dropped_train_X, dropped_val_X, dropped_test_X], ignore_index=True)
```

```python
all_X
```

```python
scaler = MinMaxScaler()
```

```python
all_X[['Fare', 'Age']] = scaler.fit_transform(all_X[['Fare', 'Age']])
```

```python
all_X
```

```python
encoded_all_X = pd.get_dummies(all_X, columns=['Sex', 'Embarked', 'title', 'cabin_level', 'ticket_prefix'])
```

```python
encoded_train_X = encoded_all_X.iloc[:dropped_train_X.shape[0]]
encoded_val_X = encoded_all_X.iloc[dropped_train_X.shape[0]: dropped_train_X.shape[0] + dropped_val_X.shape[0]]
encoded_test_X = encoded_all_X.iloc[dropped_train_X.shape[0] + dropped_val_X.shape[0]:]
```

```python
encoded_train_X
```

```python
encoded_train_X.to_csv('../data/curated/final_train_X.csv',index=False)
encoded_val_X.to_csv('../data/curated/final_val_X.csv', index=False)
encoded_test_X.to_csv('../data/curated/final_test_X.csv', index=False)
```
